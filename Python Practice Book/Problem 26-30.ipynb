## Problem 26
#### Python has built-in functions min and max to compute minimum and maximum of a given list. Provide an implementation for these functions. What happens when you call your min and max functions with a list of strings?

```python
def maxD(list) :
    max = list[0]
    for i in list :
        if max < i :
            max = i
    return max
def minD(list) :
    min = list[0]
    for i in list :
        if min > i :
            min = i
    return min    
list = [3,5,7,9,2,4,6,8,20]
#print(maxD(list))
#print(minD(list))
print("Max : {}".format(maxD(list)))
print("Min : {}".format(minD(list)))
```

```python
maxD(["a","b","c"])
```

```python
minD(["a","b","c"])
```

## Problem 27
#### Cumulative sum of a list [a, b, c, ...] is defined as [a, a+b, a+b+c, ...]. Write a function cumulative_sum to compute cumulative sum of a list. Does your implementation work for a list of strings?

```python
listA = [1,2,3,4]
listB = [4,3,2,1]
print(listA,listB)
```

```python
def cumuSum(listIn) :
    cumuValue = listIn[0]
    cumuList = []
    cumuList.append(cumuValue)
    for i in range(1, len(listIn)) :
        cumuValue = cumuValue + listIn[i]
        cumuList.append(cumuValue)
    return cumuList
```

```python
cumuSum(listA)
```

```python
cumuSum(listB)
```

## Problem 28
#### Write a function cumulative_product to compute cumulative product of a list of numbers.

```python
def cumuProduct(listIn) :
    cumuValue = listIn[0]
    cumuList = []
    cumuList.append(cumuValue)
    for i in range(1, len(listIn)) :
        cumuValue = cumuValue + listIn[i]
        cumuList.append(cumuValue)
    return cumuList
```

```python
cumuProduct(listA)
```

```python
cumuProduct(listB)
```

## Problem 29
#### Write a function unique to find all the unique elements of a list.

```python
def findUnique(listIn) :
    uniqueValue = listIn[0]
    uniqueList = []
    uniqueList.append(uniqueValue)
    for i in range(1, len(listIn)) :
        uniquecheck = 0
        uniqueValue = listIn[i]
        for j in uniqueList:
            if j == uniqueValue :
                uniquecheck = 1
        if uniquecheck == 0 :
            uniqueList.append(uniqueValue)
    return uniqueList 
```

```python
findUnique([1,2,1,3,2,5])
```

## Problem 30 
#### Write a function dups to find all duplicates in the list.

```python
def findDuplicate(listIn):
    duplicateList = []
    for i in range(len(listIn)) :
        for j in range(i+1, len(listIn)) :
            if listIn[j] == listIn[i] :
                duplicateList.append(listIn[j])
    return findUnique(duplicateList)
```

```python
findDuplicate([1,2,1,3,2,5])
```
